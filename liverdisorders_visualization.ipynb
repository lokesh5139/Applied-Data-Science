```python
# Import necessary libraries
import pandas as pd
from matplotlib import pyplot as plt  # Importing pyplot module from matplotlib for plotting
import seaborn as sns  # Importing seaborn library for enhanced data visualization
```

```python
df_liverdisorders = pd.read_csv ('/content/Indian Liver Patient Dataset (ILPD).csv')
```

```python
# @title Gender count
def gender_counts(dataframe):

    # Grouping the data by 'Gender' and obtaining the count of each group
    gender_counts = dataframe.groupby('Gender').size()

    # Plotting the grouped data as a horizontal bar chart
    ax = gender_counts.plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))

    # Iterating over each bar to annotate the counts
    for i, count in enumerate(gender_counts):
        ax.text(count, i, str(count), ha='left', va='center')  # Annotating the count on each bar

    # Getting the current Axes instance and setting the visibility of top and right spines to False for better aesthetics
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Setting labels for the axes
    plt.xlabel('Count')
    plt.ylabel('Gender')

    # Displaying the plot
    plt.show()

gender_counts(df_liverdisorders)
```

```python
# @title TB vs DB
def TBvsDB(dataframe, x_col, y_col, s=32, alpha=0.8):

    ax = dataframe.plot(kind='scatter', x=x_col, y=y_col, s=s, alpha=alpha)

    # Getting the current Axes instance and setting the visibility of top and right spines to False for better aesthetics
    ax.spines[['top', 'right']].set_visible(False)

    # Displaying the plot
    plt.show()

TBvsDB(df_liverdisorders, 'TB', 'DB', s=32, alpha=0.8)
```

```python
# @title Gender vs TB


def genderTB(dataframe, x_col, y_col):

    figsize = (12, 1.2 * len(dataframe[y_col].unique()))
    plt.figure(figsize=figsize)
    sns.violinplot(data=dataframe, x=x_col, y=y_col, inner='box', palette='Dark2')
    sns.despine(top=True, right=True, bottom=True, left=True)
    plt.show()

genderTB(df_liverdisorders, 'TB', 'Gender')
```

```python
# @title Correlation matrix

def correlation_heatmap(dataframe):

    # Compute the correlation matrix
    correlation_matrix = dataframe.corr()

    # Set up the matplotlib figure
    plt.figure(figsize=(10, 8))

    # Plot the heatmap
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

    # Add title
    plt.title('Correlation Heatmap')

    # Display the heatmap
    plt.show()

correlation_heatmap(df_liverdisorders)
```
